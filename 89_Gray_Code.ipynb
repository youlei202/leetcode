```python
from typing import List
```

```python
class Solution:
    def grayCodeBin(self, n: int) -> List[int]:
        if n == 1:
            return ["0", "1"]
        g = self.grayCodeBin(n-1)
        gg1 = list(map(lambda x: "0"+x, g))
        gg2 = list(map(lambda x: "1"+x, g))

        return gg1 + list(reversed(gg2))

    def grayCode(self, n: int) -> List[int]:
        if n < 1:
            return [0]
        bin_list = self.grayCodeBin(n)
        dec_list = list(map(lambda x: int(x, 2), bin_list))
        return dec_list

```
