```python
class Solution(object):
    def romanToInt(self, s):
        """
        :type s: str
        :rtype: int
        """
        dic = []
        dic.append({'':0, 'I':1, 'II':2, 'III':3, 'IV':4, 'V':5, 'VI':6, 'VII':7, 'VIII':8, 'IX':9})
        dic.append({'':0, 'X':1, 'XX':2, 'XXX':3, 'XL':4, 'L':5, 'LX':6, 'LXX':7, 'LXXX':8, 'XC':9})
        dic.append({'':0, 'C':1, 'CC':2, 'CCC':3, 'CD':4, 'D':5, 'DC':6, 'DCC':7, 'DCCC':8, 'CM':9})
        dic.append({'':0, 'M':1, 'MM':2, 'MMM':3})
        
        num = 0
        i = 0
        while i < len(s):
            if s[i] in ['M']: k = 3
            elif s[i] in ['C','D']: k = 2
            elif s[i] in ['X', 'L']: k = 1
            else: k = 0
            j = i
            while j+1 <= len(s) and s[i:j+1] in dic[k]: 
                j += 1
            num += dic[k][s[i:j]] * (10**k)
            i = j
                    
        return num
            
                
                
```
