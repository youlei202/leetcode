```python
class Solution(object):
    def threeSum(self, nums):
        """
        :type nums: List[int]
        :rtype: List[List[int]]
        """
        exist = {nums[i]:i for i in range(len(nums))}
        dic = {}
        
        aa = True
        for i in range(len(nums)):
            if nums[i]!=nums[0]: aa = False
        if aa == True and nums[0] == 0: 
            return [[0,0,0]]
            
        ans = []
        for i in range(len(nums)-1):
            split = len(nums)-1
            for j in range(i+1,len(nums)):
                x = -(nums[i]+nums[j])
                if exist.has_key(x):
                    a = sorted([nums[i],nums[j],x])
                    if not dic.has_key((a[0],a[1],a[2])) and exist[x]!=i and exist[x]!=j:
                        ans.append(a)
                        dic[(a[0],a[1],a[2])] = 1 

        return ans
```
