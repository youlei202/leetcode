```python
from typing import List
```

```python
class Solution:
    def eraseOverlapIntervals(self, intervals: List[List[int]]) -> int:
        if len(intervals) <= 1:
            return 0
        intervals = sorted(intervals, key = lambda x: x[1])
        count = 0
        j = 0
        for i in range(1, len(intervals)):
            if intervals[i][0] < intervals[j][1]:
                count += 1
            else:
                j = i
        return count

```
